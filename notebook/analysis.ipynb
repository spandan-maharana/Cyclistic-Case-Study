```python
#1. Library Imports
import pandas as pd
import glob
import os
```

```python
#2. Load CSV Files
path = "/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/data/*.csv"
files = sorted(glob.glob(path))
print(files)
```

```python
#3. Combining Datasets
data = []
for i in files:
    print("Loading: ", i)
    temp_df = pd.read_csv(i)
    data.append(temp_df)

df = pd.concat(data, ignore_index=True)
print("Dataset Shape: ", df.shape)
```

```python
#4. Dataset Inspection
df.info()
#df.head()
df["member_casual"].value_counts()
```

```python
#5. Converting Time Columns to Datetime
df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])
```

```python
#6. Ride Duration Feature Creation
df["ride_length"] = df["ended_at"] - df["started_at"]
df["ride_length_min"] = df["ride_length"].dt.total_seconds() / 60 # Convert to minutes
df["ride_length"].describe()
```

```python
#7. Removing Invalid Rides and Duplicates
df = df[df["ride_length_min"] > 0] # Remove rides with negative or zero duration
df = df.drop_duplicates() # Remove duplicate rows
print("Dataset Shape after cleaning: ", df.shape)
df.head()
```

```python
#8. Feature Engineering - Extracting Date Components
df["day_of_week"] = df["started_at"].dt.day_name()
df["month"] = df["started_at"].dt.month_name()
df["hour"] = df["started_at"].dt.hour
df.head()
```

```python
#9. Data Validation
df["ride_length_min"].describe()
df.isnull().sum()
df["member_casual"].value_counts()
```

```python
#10. Exporting Cleaned Dataset
output_path = "/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/data/cleaned_cyclistic_data.csv"
df.to_csv(output_path, index=False)
```
