```python
import pandas as pd

rides_user = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/rides_by_user_type.csv")
avg_length = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/avg_ride_length.csv")
weekday = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/rides_by_weekday.csv")
month = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/rides_by_month.csv")
hour = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/rides_by_hour.csv")
duration = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/ride_length_distribution.csv")
bike = pd.read_csv("/Users/spandanmaharana/Desktop/BI Case Studies/Case Study 1/Cyclistic Case Study/Power BI datasets/bike_type_usage.csv")
```

```python
df = weekday.merge(month, on="member_casual", how="left")
df = df.merge(hour, on="member_casual", how="left")

df = df.merge(rides_user, on="member_casual", how="left")
df = df.merge(avg_length, on="member_casual", how="left")
```
